#**Cronbach’s Alpha Function & Item Total Correlation Function (Pearson Product Moment Correlation)**

```python
def cronbach_alpha(df_scale: pd.DataFrame) -> float:
    """
    Hitung Cronbach's alpha untuk sekumpulan item (kolom).
    """
    itemscores = df_scale.to_numpy(dtype=float)
    n_items = itemscores.shape[1]
    item_variances = itemscores.var(axis=0, ddof=1)
    total_scores = itemscores.sum(axis=1)
    total_variance = total_scores.var(ddof=1)

    alpha = (n_items / (n_items - 1)) * (1 - item_variances.sum() / total_variance)
    return alpha


def item_total_correlation(df_scale: pd.DataFrame) -> pd.DataFrame:
    """
    Hitung corrected item-total correlation:
    korelasi tiap item dengan total skala tanpa item itu sendiri.
    """
    res = []
    for col in df_scale.columns:
        total_without = df_scale.drop(columns=[col]).sum(axis=1)
        r, p = stats.pearsonr(df_scale[col], total_without)
        res.append({
            "item": col,
            "r_item_total": r,
            "p_value": p
        })
    return pd.DataFrame(res)
```

#**Uji validitas (per konstruk X1, X2, X3, X5, X6, X7) dengan Pearson Product Moment Correlation**

```python
scales = {
    "X1": X1_cols,
    "X2": X2_cols,
    "X3": X3_cols,
    "X5": X5_cols,
    "X6": X6_cols,
    "X7": X7_cols
}

validity_results = {}

for nama, cols in scales.items():
    df_scale = dataset[cols]
    res = item_total_correlation(df_scale)
    validity_results[nama] = res

    print(f"\n=== Uji Validitas {nama} ===")
    print(res)

    print("\nItem valid (r_item_total >= 0.30 & p < 0.05):")
    print(res[(res["r_item_total"] >= 0.30) & (res["p_value"] < 0.05)])
```

#**Uji reliabilitas (Cronbach’s Alpha)**

```python
for nama, cols in scales.items():
    df_scale = dataset[cols]

    alpha = cronbach_alpha(df_scale)
    print(f"\n=== Uji Reliabilitas {nama} ===")
    print(f"Cronbach's Alpha = {alpha:.3f}")

    # Cek alpha jika item dihapus
    info = []
    for c in cols:
        alpha_if_deleted = cronbach_alpha(df_scale.drop(columns=[c]))
        info.append({
            "item": c,
            "alpha_if_deleted": alpha_if_deleted
        })
    info_df = pd.DataFrame(info)
    print("Cronbach's Alpha jika item dihapus:")
    print(info_df)
```

#**Analisis CFA (Confirmatory Factor Analysis)**

##**Menyiapkan data untuk CFA**

```python
rename_dict = {}

# X1: Pengetahuan
for i, col in enumerate(X1_cols, start=1):
    rename_dict[col] = f"KN{i}"

# X2: Risiko
for i, col in enumerate(X2_cols, start=1):
    rename_dict[col] = f"RS{i}"

# X3: Kepercayaan
for i, col in enumerate(X3_cols, start=1):
    rename_dict[col] = f"TR{i}"

# X5: Literasi
for i, col in enumerate(X5_cols, start=1):
    rename_dict[col] = f"LT{i}"

# X6: Pengaruh sosial
for i, col in enumerate(X6_cols, start=1):
    rename_dict[col] = f"SO{i}"

# X7: Minat
for i, col in enumerate(X7_cols, start=1):
    rename_dict[col] = f"IN{i}"

data_cfa = dataset[list(rename_dict.keys())].rename(columns=rename_dict)

print(data_cfa.head())
```

##**Normalization with Z-Score**

```python
data_cfa_z = (data_cfa - data_cfa.mean()) / data_cfa.std(ddof=0)
```

```python
data_cfa_z
```

##**Mendefinisikan Spesifikasi Model Pengukuran**

```python
model_desc = """
# Faktor laten & indikatornya
Knowledge =~ KN1 + KN2 + KN3 + KN4 + KN5
Risk      =~ RS1 + RS2 + RS3 + RS4 + RS5
Trust     =~ TR1 + TR2 + TR3 + TR4 + TR5
Literacy  =~ LT1 + LT2 + LT3 + LT4
Social    =~ SO1 + SO2 + SO3 + SO4 + SO5
Intention =~ IN1 + IN2 + IN3 + IN4 + IN5

# Korelasi antar faktor (sesuai teori, di sini semua dibiarkan saling berkorelasi)
Knowledge ~~ Risk + Trust + Literacy + Social + Intention
Risk      ~~ Trust + Literacy + Social + Intention
Trust     ~~ Literacy + Social + Intention
Literacy  ~~ Social + Intention
Social    ~~ Intention
"""
```

##**Fitting CFA**

```python
model = Model(model_desc)

res = model.fit(data_cfa_z)
print(res)
```

##**Evaluasi CFA**

###**Signifikansi Factor loadings**

```python
params = model.inspect()

# factor loading (item <- faktor)
loadings = params[params['op'] == '~']
print("\nFACTOR LOADINGS:")
print(loadings)
```

###**Goodness-of-fit index**

```python
stats_fit = calc_stats(model)

print("Chi-square:", stats_fit.chi2.iloc[0])
print("df:", stats_fit['DoF'].iloc[0])
print("p-value:", stats_fit['chi2 p-value'].iloc[0])
print("CFI:", stats_fit['CFI'].iloc[0])
print("TLI:", stats_fit['TLI'].iloc[0])
print("RMSEA:", stats_fit['RMSEA'].iloc[0])
```
